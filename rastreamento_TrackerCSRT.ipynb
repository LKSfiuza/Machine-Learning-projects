```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
video = cv2.VideoCapture('/content/rua.mp4')
```

```python
rastreador = cv2.TrackerCSRT_create()
```

```python
ok, frame = video.read()
```

```python
#bbox = cv2.selectROI("Frame", frame, False)
bbox = (560, 210, 170, 160)
```

```python
ok = rastreador.init(frame, bbox)
```

```python
while True:
  ok, frame = video.read()
  if not ok:
    break
  ok, bbox = rastreador.update(frame)
  #print(bbox)
  if ok:
    (x,y,w,h) = [int(v) for v in bbox]
    cv2.rectangle(frame, (x,y), (x+w, y+h),(0,255,0),2,1)
  else:
    cv2.putText(frame, "ERRO",(100,80),cv2.FONT_HERSHEY_SIMPLEX, .75,(0,0,255),2)
  cv2_imshow(frame)
  if cv2.waitKey(1) & 0XFF == 27:
    break
```
