# DBSCAN

## Base Salario x Idade

```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
x = [20, 27, 21, 37, 46, 53,55,47,52,32,39,41,39,48,48]
y = [1000, 1200, 2900,1850,900,950,2000,2100,3000,5900,4100,5100,7000,5000,6500]
```

```python
base_salario = np.array([[20,1000], [27, 1200], [21, 2900], [37, 1850], [46, 900], [53, 950], [55, 2000], [47, 2100], [52, 3000], [32, 5900], [39, 4100], [41, 5100], [39, 7000], [48, 5000], [48, 6500]])
base_salario
```

```python
scaler_salario = StandardScaler()
base_salario = scaler_salario.fit_transform(base_salario)
base_salario
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan_salario = DBSCAN(eps=0.95, min_samples=2)
dbscan_salario.fit(base_salario)
```

```python
rotulos = dbscan_salario.labels_
rotulos
```

```python
graficos = px.scatter(x=x, y=y, color=rotulos)
graficos.show()
```

## Base cartão de crédito

```python
base_cartao = pd.read_csv('credit_card_clients.csv', header=1)
base_cartao
```

```python
base_cartao['BILL_TOT'] = base_cartao['BILL_AMT1'] + base_cartao['BILL_AMT2'] + base_cartao['BILL_AMT3'] + base_cartao['BILL_AMT4'] + base_cartao['BILL_AMT5']
```

```python
X_cartao = base_cartao.iloc[:,[1,25]].values
X_cartao
```

```python
scaler_cartao = StandardScaler()
X_cartao = scaler_cartao.fit_transform(X_cartao)
X_cartao
```

```python
dbscan_cartao = DBSCAN(eps=0.37, min_samples=5)
rotulos = dbscan_cartao.fit_predict(X_cartao)
rotulos
```

```python
np.unique(rotulos, return_counts=True)
```

```python
graficos = px.scatter(x=X_cartao[:,0], y=X_cartao[:,1], color=rotulos)
graficos.show()
```

## DBSCAN x Hierárquico x K-means

```python
from sklearn import datasets
```

```python
X_random, y_random = datasets.make_moons(n_samples=1500, noise=0.09)
```

```python
grafico = px.scatter(x=X_random[:,0], y=X_random[:,1])
grafico.show()
```

## KMEANS

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(X_random)
```

```python
grafico = px.scatter(x=X_random[:,0], y=X_random[:,1], color=kmeans.labels_)
grafico.show()
```

## HIERARQUICO

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
rotulos = hc.fit_predict(X_random)
grafico = px.scatter(x=X_random[:,0], y=X_random[:,1], color=rotulos)
grafico.show()
```

## DBSCAN

```python
dbscan = DBSCAN(eps=0.1)
rotulos = dbscan.fit_predict(X_random)
grafico = px.scatter(x=X_random[:,0], y=X_random[:,1], color=rotulos)
grafico.show()
```
