# Redes Neurais Artificiais


```python
from sklearn.neural_network import MLPClassifier
```

## Base de dados de credit

```python
import pickle
with open('credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento , X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape,y_credit_treinamento.shape
```

```python
X_credit_teste.shape,y_credit_teste.shape
```

```python
RN_credit = MLPClassifier(max_iter=1500, verbose=True,tol=0.0000100,
                          activation='relu',hidden_layer_sizes=(2,2))
RN_credit.fit(X_credit_treinamento,y_credit_treinamento)
```

```python
previsoes = RN_credit.predict(X_credit_teste)
previsoes
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_credit_teste,previsoes)
```

```python
print(classification_report(y_credit_teste,previsoes))
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(RN_credit)
cm.fit(X_credit_treinamento,y_credit_treinamento)
cm.score(X_credit_teste,y_credit_teste)
```
