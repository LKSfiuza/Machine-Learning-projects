# K-means - Agrupamento

```python
!pip install plotly --upgrade
```

```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
x = [20, 27, 21, 37, 46, 53,55,47,52,32,39,41,39,48,48]
y = [1000, 1200, 2900,1850,900,950,2000,2100,3000,5900,4100,5100,7000,5000,6500]
```

```python
grafico =px.scatter(x=x, y=y, title='Agrupamento de Clientes')
grafico.show()
```

```python
matriz = []
for i in range(len(x)):
  matriz.append([x[i], y[i]])
matriz
```

```python
base_salario = np.array([[20,1000], [27, 1200], [21, 2900], [37, 1850], [46, 900], [53, 950], [55, 2000], [47, 2100], [52, 3000], [32, 5900], [39, 4100], [41, 5100], [39, 7000], [48, 5000], [48, 6500]])
base_salario
```

```python
scaler_salario = StandardScaler()
base_salario = scaler_salario.fit_transform(base_salario)
base_salario
```

```python
matriz = scaler_salario.fit_transform(matriz)
matriz
```

```python
kmeans_salario = KMeans(n_clusters=3)
kmeans_salario.fit(base_salario)
```

```python
c = kmeans_salario.cluster_centers_
c
```

```python
scaler_salario.inverse_transform(c)
```

```python
r = kmeans_salario.labels_
```

```python
r
```

```python
grafico1 = px.scatter(x=base_salario[:,0], y=base_salario[:,1], color=r)
grafico2 = px.scatter(x=c[:,0], y=c[:,1], size=[12, 12, 12])
grafico3 = go.Figure(data=grafico1.data + grafico2.data)
grafico3.show()
```

# Dados Randômicos

```python
from sklearn.datasets import make_blobs
```

```python
X_random, y_random = make_blobs(n_samples=200, centers=5, random_state=0)
```

```python
X_random
```

```python
y_random
```

```python
grafico = px.scatter(x=X_random[:,0], y=X_random[:,1])
grafico.show()
```

```python
kmeans_blobs = KMeans(n_clusters=5)
kmeans_blobs.fit(X_random)
```

```python
r1 = kmeans_blobs.predict(X_random)
r1
```

```python
c1 = kmeans_blobs.cluster_centers_
c1
```

```python
grafico1 = px.scatter(x=X_random[:,0], y=X_random[:,1], color=r1)
grafico2 = px.scatter(x=c1[:,0], y=c1[:,1], size=[12, 12, 12,12,12])
grafico3 = go.Figure(data=grafico1.data + grafico2.data)
grafico3.show()
```

# Base Cartão de Crédito

## Atributos limitados


```python
base_cartao = pd.read_csv('credit_card_clients.csv', header=1)
base_cartao
```

```python
base_cartao['BILL_TOT'] = base_cartao['BILL_AMT1'] + base_cartao['BILL_AMT2'] + base_cartao['BILL_AMT3'] + base_cartao['BILL_AMT4'] + base_cartao['BILL_AMT5']
```

```python
base_cartao
```

```python
X_cartao = base_cartao.iloc[:,[1,25]].values
X_cartao
```

```python
scaler_cartao = StandardScaler()
X_cartao = scaler_cartao.fit_transform(X_cartao)
X_cartao
```

```python
wcss = []
for i in range(1,11):
  kmeans_cartao = KMeans(n_clusters=i, random_state=0)
  kmeans_cartao.fit(X_cartao)
  wcss.append(kmeans_cartao.inertia_)
```

```python
wcss
```

```python
grafico = px.line(x=range(1,11), y=wcss)
grafico.show()
```

```python
kmeans_cartao = KMeans(n_clusters=4, random_state=0)
kmeans_cartao.fit(X_cartao)
```

```python
rotulos = kmeans_cartao.fit_predict(X_cartao)
rotulos
```

```python
grafico1 = px.scatter(x=X_cartao[:,0], y=X_cartao[:,1], color=rotulos)
grafico1
```

```python
lista_clientes = np.column_stack((base_cartao, rotulos))
lista_clientes
```

```python
lista_clientes = lista_clientes[lista_clientes[:,26].argsort()]
lista_clientes
```
