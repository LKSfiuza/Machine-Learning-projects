```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from yellowbrick.regressor import ResidualsPlot
```

```python
base_plano_saude2 = pd.read_csv('plano_saude2.csv')
base_plano_saude2.head()
```

```python
X_plano_saude2 = base_plano_saude2.iloc[:, 0:1].values
y_plano_saude2 = base_plano_saude2.iloc[:, 1].values
```

```python
X_plano_saude2
```

```python
y_plano_saude2
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor_arvore_saude = DecisionTreeRegressor()
regressor_arvore_saude.fit(X_plano_saude2, y_plano_saude2)
```

```python
previsoes = regressor_arvore_saude.predict(X_plano_saude2)
previsoes
```

```python
regressor_arvore_saude.score(X_plano_saude2, y_plano_saude2)
```

```python
grafico = px.scatter(x = X_plano_saude2.ravel(), y = y_plano_saude2)
grafico.add_scatter(x = X_plano_saude2.ravel(), y = previsoes, name = 'Regressão')
grafico.show()
```

```python
X_teste_arvore = np.arange(min(X_plano_saude2), max(X_plano_saude2), 0.1)
X_teste_arvore
```

```python
X_teste_arvore = X_teste_arvore.reshape(-1, 1)
X_teste_arvore.shape
```

```python
grafico = px.scatter(x = X_plano_saude2.ravel(), y = y_plano_saude2)
grafico.add_scatter(x = X_teste_arvore.ravel(), y = regressor_arvore_saude.predict(X_teste_arvore), name = 'Regressão')
grafico.show()
```

```python
regressor_arvore_saude.predict([[40]])
```

```python
regressor_arvore_saude.predict([[40.6]])
```

##Base preço das casas

```python
base_pcasas = pd.read_csv('house_prices.csv')
base_pcasas.head()
```

```python
X_casas = base_pcasas.iloc[:, 3:19].values
y_casas = base_pcasas.iloc[:, 2].values
```

```python
from sklearn.model_selection import train_test_split
X_casas_treinamento, X_casas_teste,y_casas_treinamentro, y_casas_teste = train_test_split(X_casas, y_casas, test_size = 0.3, random_state = 0)
```

```python
X_casas_treinamento.shape
```

```python
y_casas_treinamentro
```

```python
X_casas_teste.shape
```

```python
regressor_arvore_casas = DecisionTreeRegressor()
regressor_arvore_casas.fit(X_casas_treinamento, y_casas_treinamentro)
```

```python
regressor_arvore_casas.score(X_casas_treinamento, y_casas_treinamentro)
```

```python
regressor_arvore_casas.score(X_casas_teste, y_casas_teste)
```

```python
previsoes_casas = regressor_arvore_casas.predict(X_casas_teste)
previsoes_casas
```

```python
y_casas_teste
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_casas_teste, previsoes_casas)
```
