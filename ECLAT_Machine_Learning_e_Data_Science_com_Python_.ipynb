```python
base_mercado1 = pd.read_csv('mercado.csv', header = None)
```

```python
eclat.df_bin
```

```python
eclat.uniq_
```

```python
indices, suporte = eclat.fit(min_support=0.3, min_combination= 1 , max_combination = 3)
```
