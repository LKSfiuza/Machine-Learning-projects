```python
import pandas as pd
from apyori import apriori
```

## BASE DE DADOS MERCADO 1

```python
base_mercado1 = pd.read_csv('mercado.csv', header = None)
base_mercado1
```

```python
transacoes = []
for i in range(len(base_mercado1)):
  #print(base_mercado1.values[i, 0])
  transacoes.append([str(base_mercado1.values[i, j]) for j in range(base_mercado1.shape[1])])
```

```python
transacoes
```

```python
type(transacoes)
```

```python
transacoes2 = []
transacoes2 = base_mercado1.apply(lambda row: row.astype(str).tolist(), axis=1).tolist()
transacoes2
```

```python
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.8, min_lift = 2)
resultados = list(regras)
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []
for resultado in resultados:
  s = resultado[1]
  regras = resultado[2]
  for regra in regras:
    a = list(regra[0])
    b = list(regra[1])
    c = regra[2]
    l = regra[3]
    print(a, ' - ', b, ' - ', c, ' - ', l)
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

## BASE DE DADOS MERCADO 2

```python
base_mercado2 = pd.read_csv('mercado2.csv', header=None)
base_mercado2
```

```python
transacoes2 = []
transacoes2 = base_mercado2.apply(lambda row: row.astype(str).tolist(), axis=1).tolist()
```

```python
type(transacoes2)
```

```python
regras = apriori(transacoes2, min_support = 0.003, min_confidence = 0.2, min_lift = 3)
resultados = list(regras)
len(resultados)
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []
for resultado in resultados:
  s = resultado[1]
  regras = resultado[2]
  for regra in regras:
    a = list(regra[0])
    b = list(regra[1])
    c = regra[2]
    l = regra[3]
    A.append(a)
    B.append(b)
    suporte.append(s)
    confianca.append(c)
    lift.append(l)
```

```python
regras_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
regras_df.sort_values(by = 'confianca', ascending = False)
```
